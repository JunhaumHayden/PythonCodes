```python
import numpy as np #para manipular os vetores
from matplotlib import pyplot as plt #para plotar os gr√°ficos
from sklearn.cluster import KMeans #para usar o KMeans
from sklearn.cluster import DBSCAN #para usar o KMeans
```

```python
dataset = np.array(
#matriz com as coordenadas geogr√°ficas de cada loja
[[-25, -46], #s√£o paulo
[-22, -43], #rio de janeiro
[-25, -49], #curitiba
[-30, -51], #porto alegre
[-19, -43], #belo horizonte
[-15, -47], #brasilia
[-12, -38], #salvador
[-8, -34], #recife
[-16, -49], #goiania
[-3, -60], #manaus
[-22, -47], #campinas
[-3, -38], #fortaleza
[-21, -47], #ribeir√£o preto
[-23, -51], #maringa
[-27, -48], #florian√≥polis
[-21, -43], #juiz de fora
[-1, -48], #bel√©m
[-10, -67], #rio branco
[-8, -63]]) #porto velho

plt.scatter(dataset[:,1], dataset[:,0]) #posicionamento dos eixos x e y
plt.xlim(-75, -30) #range do eixo x
plt.ylim(-50, 10) #range do eixo y
plt.grid() #fun√ß√£o que desenha a grade no nosso gr√°fico
```

# Nova se√ß√£o

```python
kmeans = KMeans(n_clusters = 2, #numero de clusters
                init = 'k-means++', n_init = 10, #algoritmo que define a posi√ß√£o dos clusters de maneira mais assertiva
                max_iter = 300) #numero m√°ximo de itera√ß√µes

```

```python
group = kmeans.fit_predict(dataset)
group
```

```python

plt.scatter(dataset[:,1], dataset[:,0], c = group) #posicionamento dos eixos x e y
plt.xlim(-75, -30) #range do eixo x
plt.ylim(-50, 10) #range do eixo y
plt.grid() #fun√ß√£o que desenha a grade no nosso gr√°fico
plt.scatter(kmeans.cluster_centers_[:,1],
            kmeans.cluster_centers_[:,0],
            s = 70, c = 'red') #posi√ß√£o de cada centroide no gr√°fico
plt.show()
```
