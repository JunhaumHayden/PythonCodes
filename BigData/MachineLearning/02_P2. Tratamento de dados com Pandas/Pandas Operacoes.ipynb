# Dados ausentes

Vamos mostrar alguns métodos convenientes para lidar com Missing Data em pandas:

```python
import numpy as np
```

```python
df = pd.DataFrame({'A':[1,2,np.nan],
                  'B':[5,np.nan,np.nan],
                  'C':[1,2,3]})
```

```python
# prompt: como remover os dados faltantes na coluna A e C

# Removendo linhas com valores faltantes na coluna 'A' e 'C'
df4 = df.dropna(subset=['A', 'C'])
df4
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
#remove a coluna - nao é muito comum
```

```python
df.dropna(thresh=2)
#indica o número de ausentes para que a linha ou coluna seja removida
```

```python
df.fillna(value='Conteúdo')
#df.fillna(value=0)
#substitui NaN por um valor específico
```

```python
df['A'].fillna(value=df['A'].mean())
#Na coluna A está substituindo os NaN por uma média da propria coluna. Boa solução para alguns casos. Não todos.
```
