```python
import pandas as pd
import numpy as np
```

```python
dados = pd.read_json('../../dataset/dados_hospedagem.json')
dados.head()
```

```python
dados = pd.json_normalize(dados['info_moveis'])
dados.head()
```

```python
colunas = list(dados.columns)
colunas
```

```python
# O Metodo explode transforma uma coluna de listas em várias linhas
dados = dados.explode(colunas[3:])
dados
```

```python
# O Metodo reset_index redefine o índice do DataFrame, removendo o índice antigo(inlace) e criando um novo índice sequencial(drop)
dados.reset_index(inplace=True, drop = True)
dados
```

```python
# O Método astype converte o tipo de dados de uma coluna para outro tipo especificado. É necessario fazer a atribuição do resultado a uma variável ou coluna, pois o método não altera o DataFrame original por padrão.
dados['max_hospedes'] = dados['max_hospedes'].astype(np.int64)
dados.info()
```

```python
col_numericas = ['quantidade_banheiros', 'quantidade_quartos', 'quantidade_camas']
col_numericas
```

```python
dados[col_numericas] = dados[col_numericas].astype(np.int64)
dados.info()
```

```python
# Lidando com colunas com preços
dados[['preco', 'taxa_deposito', 'taxa_limpeza']]
```

```python
# Função lambda para remover caracteres não numericos como por exemplo $
# para cada variavel x, trocar $ por '' e trocar ',' por '' e retirar espaços
dados['preco'] = dados['preco'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
dados['preco']
```

```python
dados['preco'].info()
```

```python
# Convertendo para float
dados['preco'] = dados['preco'].astype(np.float64)
dados.info()
```

```python
# para uma lista de colunas utiliza-se o metodo applymap().
# OBS: /tmp/ipykernel_11901/1806409663.py:1: FutureWarning: DataFrame.applymap has been deprecated. Use DataFrame.map instead.
#  dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())

dados[['taxa_deposito', 'taxa_limpeza']] = dados[['taxa_deposito', 'taxa_limpeza']].map(lambda x: x.replace('$', '').replace(',', '').strip())
dados[['taxa_deposito', 'taxa_limpeza']]
```

```python
# Tratando itens textuais
# Tokenização: Caso haja o interesse em tratar melhor as informações textuais, aplica-se a tecnica de tokenização.
# 1 - precisa-se padroniza o texto, como por exemplo, utilizar apenas caracteres em minusculos
# com  pandas, pode-se utilizar o metodo str para aplicar todos os metodos da classe str.
dados['descricao_local'] = dados['descricao_local'].str.lower()
dados['descricao_local']
```

```python
# 2 - Retirar caracteres especiais, como por exemplo, pontuação, acentos, etc.
dados['descricao_local'][3169]
```

```python
# Usando o método str.replace com regex=True, podemos remover caracteres especiais. [^\w\s] significa "qualquer coisa que não (^) seja uma letra, número ou espaço".
dados['descricao_local'].str.replace(r'[^\w\s]', '', regex=True)
```

```python
dados['descricao_local'][3169]
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace(r'[^a-zA-Z0-9\-\´]', ' ', regex=True)
dados['descricao_local']
```

```python
# 3 - Remover os hipens, ou seja, os espaços em branco. (?<!\w)-(?<!\w) significa "um hífen que não é precedido ou seguido por uma letra, número ou sublinhado".
dados['descricao_local'] = dados['descricao_local'].str.replace(r'(?<!\w)-(?<!\w)', ' ', regex=True)
dados['descricao_local']
```

```python
# 4 - Remover stopwords, ou seja, palavras que não agregam valor semântico ao texto, como artigos, preposições, etc.
```

```python
# Efetivando a Tokenização
dados['descricao_local'].str.split()
```

```python
dados['descricao_local'] = dados['descricao_local'].str.split()
dados['descricao_local']
```

```python
# tratando as demais colunas de texto
dados['comodidades'].str.replace('\{|}|\"', '', regex=True)
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"', '', regex=True)
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```
