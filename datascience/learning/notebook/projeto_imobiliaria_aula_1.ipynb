```python
imoveis_comerciais = ['Conjunto Comercial/Sala', 
                      'Prédio Inteiro', 'Loja/Salão', 
                      'Galpão/Depósito/Armazém', 
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria']
                      
# removendo imóveis comerciais
df = dados.query('@imoveis_comerciais not in Tipo')

# selecionando apenas imóveis do tipo apartamento
#df = df.query('Tipo == "Apartamento"')

# visualizando a base de dados
df.head()
```

```python
df = df.query('Suites >= 1')
df.head()
```

```python
df_suites = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_suites.plot(kind='barh', color='red', title='Valor médio do aluguel por tipo de imóvel com suítes', figsize=(10, 6), legend=False)
```

```python
import matplotlib.pyplot as plt

# Filtrar imóveis com pelo menos 1 suíte
df_suites = df[df['Suites'] >= 1]

# Agrupar por tipo e calcular quantidade e média de valor
suite_stats = df_suites.groupby('Tipo').agg(
    quantidade=('Suites', 'count'),
    media_valor=('Valor', 'mean')
).sort_values('quantidade', ascending=False)

fig, ax1 = plt.subplots(figsize=(10, 6))

# Barras: quantidade de imóveis com suíte por tipo
suite_stats['quantidade'].plot(kind='bar', color='skyblue', ax=ax1, position=0, width=0.4)
ax1.set_ylabel('Quantidade de imóveis com suíte')
ax1.set_xlabel('Tipo de imóvel')
ax1.set_title('Quantidade de imóveis com suíte e média de valor por tipo')
ax1.tick_params(axis='x', rotation=45)

# Linha: média de valor por tipo
ax2 = ax1.twinx()
suite_stats['media_valor'].plot(kind='line', color='red', marker='o', ax=ax2)
ax2.set_ylabel('Média de valor (R$)')

plt.tight_layout()
plt.show()
```

Existem diversas formas de tratar dados nulos com o Pandas. Algumas das principais formas são:

- `Remover os dados nulos`: É possível remover as linhas ou colunas que possuem valores nulos utilizando o método `dropna()` . Esse método remove todas as linhas ou colunas que possuem pelo menos um valor nulo.

- `Preencher os dados nulos`: Utilizando o método `fillna()`, podemos preencher os valores nulos com um valor específico. Além disso, também é possível utilizar argumentos específicos do método `fillna()` como o `method=”ffill”` ou `method=”bfill”` para preencher os valores nulos com o valor anterior ou posterior, respectivamente.

- `Interpolar os dados nulos`: É possível utilizar o método `interpolate()` para preencher os valores nulos com valores interpolados, ou seja, valores calculados a partir dos valores vizinhos.
